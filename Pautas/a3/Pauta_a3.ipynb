##Importación de librerías

```python
import sqlite3
```

## Conectar db a RAM

```python
conn = sqlite3.connect('backup.db')
cursor = conn.cursor()
```

## P1.

```python
cursor.execute('''
CREATE TABLE Categoria (
    id INT PRIMARY KEY,
    nombre VARCHAR(100) NOT NULL,
    descripcion TEXT
);
''')
cursor.execute('''
CREATE TABLE Proveedor (
    rut INT PRIMARY KEY,
    nombre VARCHAR(100) NOT NULL,
    web VARCHAR(100),
    telefono VARCHAR(20),
    calle VARCHAR(100),
    numero INT,
    comuna VARCHAR(50),
    ciudad VARCHAR(50)
);
''')
cursor.execute('''
CREATE TABLE Producto (
    id INT PRIMARY KEY,
    rut_proveedor INT,
    id_categoria INT,
    nombre VARCHAR(100) NOT NULL,
    precio_actual INT,
    stock INT,
    FOREIGN KEY (rut_proveedor) REFERENCES Proveedor(rut),
    FOREIGN KEY (id_categoria) REFERENCES Categoria(id)
);
''')
cursor.execute('''
CREATE TABLE Cliente (
    rut INT PRIMARY KEY,
    nombre VARCHAR(100) NOT NULL,
    calle VARCHAR(100),
    numero INT,
    comuna VARCHAR(50),
    ciudad VARCHAR(50)
);
''')
cursor.execute('''
CREATE TABLE Telefono (
    telefono VARCHAR(20),
    rut_cliente INT,
    PRIMARY KEY (telefono, rut_cliente),
    FOREIGN KEY (rut_cliente) REFERENCES Cliente(rut)
);
''')
cursor.execute('''
CREATE TABLE Venta (
    id INT PRIMARY KEY,
    rut_cliente INT,
    fecha DATE,
    descuento DECIMAL(10, 2),
    FOREIGN KEY (rut_cliente) REFERENCES Cliente(rut)
);
''')
cursor.execute('''
CREATE TABLE Venta_Producto (
    id_venta INT,
    id_producto INT,
    precio_venta INT,
    cantidad INT,
    PRIMARY KEY (id_venta, id_producto),
    FOREIGN KEY (id_venta) REFERENCES Venta(id),
    FOREIGN KEY (id_producto) REFERENCES Producto(id)
);
''')
```

## P3.

```python
cursor.execute('''
CREATE TABLE DEPARTAMENTO (
    codigo_departamento INT PRIMARY KEY,
    nombre VARCHAR(100) NOT NULL
);
''')
cursor.execute('''
CREATE TABLE CONFERENCIA (
    codigo_conferencia INT PRIMARY KEY,
    nombre VARCHAR(150) NOT NULL,
    ciudad VARCHAR(100),
    fecha DATE,
    codigo_departamento INT,
    codigo_articulo INT,
    FOREIGN KEY (codigo_departamento) REFERENCES DEPARTAMENTO(codigo_departamento),
    FOREIGN KEY (codigo_articulo) REFERENCES ARTICULO(codigo_articulo)
);
''')
cursor.execute('''
CREATE TABLE INVESTIGADOR (
    rut INT PRIMARY KEY,
    nombre_completo VARCHAR(150) NOT NULL,
    correo VARCHAR(100) UNIQUE
);
''')
cursor.execute('''
CREATE TABLE INVESTIGADOR_CONFERENCIA (
    codigo_conferencia INT,
    rut INT,
    fecha_inscripcion DATE,
    tipo_participacion VARCHAR(50) CHECK (tipo_participacion IN ('asistente', 'expositor', 'organizador')),
    PRIMARY KEY (codigo_conferencia, rut),
    FOREIGN KEY (codigo_conferencia) REFERENCES CONFERENCIA(codigo_conferencia),
    FOREIGN KEY (rut) REFERENCES INVESTIGADOR(rut)
);
''')
cursor.execute('''
CREATE TABLE ARTICULO (
    codigo_articulo INT PRIMARY KEY,
    titulo VARCHAR(200) NOT NULL,
    resumen TEXT,
    fecha_envio DATE
);
''')
cursor.execute('''
CREATE TABLE INVESTIGADOR_ARTICULO (
    codigo_articulo INT,
    rut INT,
    PRIMARY KEY (codigo_articulo, rut),
    FOREIGN KEY (codigo_articulo) REFERENCES ARTICULO(codigo_articulo),
    FOREIGN KEY (rut) REFERENCES INVESTIGADOR(rut)
);
''')
```

- Nótese que todas las tablas se guardan en el mismo archivo. Si quieres guardarlas en archivos diferentes solo debes crear otro cursor que apunte a una db distinta.

## Liberar conexión

```python
conn.close() #Esto libera la memoria (cuando se guarda la bd en RAM).
```
