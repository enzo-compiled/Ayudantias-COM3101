#Creación DB

##Importación de librerías

```python
import sqlite3
```

## Conectar db a RAM

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

* `sqlite3.connect(':memory:')` abre una base de datos temporal que vive solo en la memoria RAM.
* `conn` es el objeto de conexión, que usarás para comunicarte con la base.
* `conn.cursor()` te da un cursor, que sirve para ejecutar comandos SQL.

Si hubieramos escrito `sqlite3.connect('nombre.db')`, se crearía un archivo persistente en disco.

## Creación simple de tabla

- Considere el modelo: STUDENT(Ssn PK, Name, Home_phone, Address, Office_phone, Age, Gpa)

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS Student(
  Ssn INT PRIMARY KEY,
  Name VARCHAR(30),
  Home_Phone VARCHAR(20),
  Address VARCHAR(40) NOT NULL,
  Office_Phone VARCHAR(20),
  Age INT NOT NULL,
  GPA REAL
  );
''')
```

## Liberar conexión

```python
conn.close() #Esto libera la memoria.
```
